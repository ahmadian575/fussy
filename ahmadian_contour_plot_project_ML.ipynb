# Import libraries
we need to matplotlab and numpy

```python
from matplotlib import pyplot as plt
import numpy as np #For mathematics, and making arrays
```

# Create a panel

```python
width_panel = 4
height_panel = 3
d = 1000
plt.figure(figsize=(width_panel, height_panel), dpi=d)
```

# Introduce the data
This usually comprises of two independent-variable arrays and a dependent variable array (the contour).

```python
x = np.arange(0, 25, 1)
y = np.arange(0, 25, 1)
x, y = np.meshgrid(x, y)
z = np.sin(x/3) + np.cos(y/4)
```

# Draw the contour plot

```python
plt.contour(x, y, z)
```

# Make the contour plot
Now, lets combine all of this to make a simple contour plot.

```python
import matplotlib.pyplot as plt
import numpy as np

width_panel = 4
height_panel = 3
d = 1000
plt.figure(figsize=(width_panel, height_panel), dpi=d)
x = np.arange(0, 25, 1)
y = np.arange(0, 25, 1)
x, y = np.meshgrid(x, y)
z = np.sin(x/3) + np.cos(y/4)#np.sin(x * np.pi / 2) + np.cos(y * np.pi / 3)

plt.contour(x, y, z)
```

# Fill in the contour plot
To make a filled contour plot, just replace contour , contourf.

```python
import matplotlib.pyplot as plt
import numpy as np

width_panel = 4
height_panel = 3
d = 1000
plt.figure(figsize=(width_panel, height_panel), dpi=d)
x = np.arange(0, 25, 1)
y = np.arange(0, 25, 1)
x, y = np.meshgrid(x, y)
z = np.sin(x/3) + np.cos(y/4)#np.sin(x * np.pi / 2) + np.cos(y * np.pi / 3)

plt.contourf(x, y, z)
```
